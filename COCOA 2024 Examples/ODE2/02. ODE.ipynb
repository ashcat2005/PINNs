## First Order ODE solution (Exponential Decay)

Considera the first order ODE (exponential decay):
$$
\frac{d u}{d t} = - \gamma u 
$$
where $\gamma = 2$ is a constant parameter. 

In order to solve the system, we consider the initial condition
$$
u(t=0) = u_0 = 10
$$
and we will solve in the interval $t\in [0,10]$.

The known analytic solution of this problem is: $u(t) = u_0 e^{-\gamma t}$. 

```python
import pinnde.ode_Solvers as ode_Solvers
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

eqn = "ut + 2*u"
order = 1
inits = [10]
t_bdry = [0,10]
N_pde = 100
epochs = 1000

mymodel = ode_Solvers.solveODE_IVP(eqn, order, inits, t_bdry, N_pde, epochs)

mymodel.plot_epoch_loss('epoch_loss.png')

mymodel.plot_solution_prediction('solution.png')

exact_eqn = "10*np.exp(-2*t)"
mymodel.plot_predicted_exact(exact_eqn, 'comparison.png')
```
