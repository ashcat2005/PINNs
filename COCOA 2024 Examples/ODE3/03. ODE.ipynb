## Second Order ODE  

$$
\frac{d^2 u}{dt^2} = \left(\frac{du}{dt}\right)^2
$$

Initial condition: 
$$
\begin{cases}
u(t=0) = & 1\\
u'(t=0) = & 0.5
\end{cases}
$$

```python
import pinnde.ode_Solvers as ode_Solvers
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


eqn = "utt - ut**2"
order = 2
inits = [1, 0.5]
t_bdry = [0,1]
N_pde = 100
epochs = 1000

mymodel = ode_Solvers.solveODE_IVP(eqn, order, inits, t_bdry, N_pde, epochs)


mymodel.plot_epoch_loss('epoch_loss.png')

mymodel.plot_solution_prediction('solution.png')

exact_eqn = "-(np.log(abs(t-2))) + np.log(2) + 1"
mymodel.plot_predicted_exact(exact_eqn, 'comparison.png')
```

```python
plt.imshow(mpimg.imread('epoch_loss.png'))
plt.axis('off')
plt.show()
```
