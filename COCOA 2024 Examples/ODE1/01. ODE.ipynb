## Second Order ODE solution

Considera the second order ODE:
$$
\frac{d^2 u}{d t^2} = - \omega^2 u 
$$
where $\omega^2 = 2$ is a constant parameter. 

In order to solve the system, we consider the initial condition
$$
\begin{cases}
u(t=0) & = u_0 = 1\\
u'(t=0)   & = 0
\end{cases}
$$
and we will solve in the interval $t\in [0,2\pi]$.

The known analytic solution of this problem is: $u(t) = u_0 \cos (\sqrt{2}t)$. 

```python
import pinnde.ode_Solvers as ode_Solvers
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

eqn = "utt + 2*u"
order = 2
inits = [1, 0.]
t_bdry = [0,2*np.pi]
N_pde = 100
epochs = 1000

mymodel = ode_Solvers.solveODE_IVP(eqn, order, inits, t_bdry, N_pde, epochs)

mymodel.plot_epoch_loss('epoch_loss.png')

mymodel.plot_solution_prediction('solution.png')

exact_eqn = "np.cos(np.sqrt(2)*t)"
mymodel.plot_predicted_exact(exact_eqn, 'comparison.png')
```
