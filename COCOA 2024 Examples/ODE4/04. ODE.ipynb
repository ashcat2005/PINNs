## Third Order ODE solution 

$$
\frac{d^3 u}{dt^3} -2 \frac{d^2 u}{dt^2} + u = 0
$$

Initial condition: 
$$
\begin{cases}
u(t=0) = & 0\\
u'(t=0) = & 1\\
u''(t=0) = & 0
\end{cases}
$$

```python
import pinnde.ode_Solvers as ode_Solvers
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt


eqn = "uttt - 2*utt + u"
order = 3
inits = [0, 1, 0]
t_bdry = [0,1]
N_pde = 200
epochs = 1000

mymodel = ode_Solvers.solveODE_IVP(eqn, order, inits, t_bdry, N_pde, epochs, constraint = "hard")


mymodel.plot_epoch_loss('epoch_loss.png')

mymodel.plot_solution_prediction('solution.png')
```

```python
plt.imshow(mpimg.imread('epoch_loss.png'))
plt.axis('off')
plt.show()
```
